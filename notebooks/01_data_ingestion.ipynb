```python
BASE = Path().resolve().parent  
fund_master = pd.read_csv(BASE / "data" / "raw" / "01_fund_master.csv")

print("=== FUND HOUSES ===")
print(fund_master["fund_house"].unique())
print(f"Total unique fund houses: {fund_master['fund_house'].nunique()}")

print("\n=== CATEGORIES ===")
print(fund_master["category"].value_counts())

print("\n=== SUB-CATEGORIES ===")
print(fund_master["sub_category"].value_counts())

print("\n=== RISK GRADES ===")
print(fund_master["risk_category"].value_counts())

print("\n=== SEBI CATEGORY CODES ===")
print(fund_master["sebi_category_code"].unique())
```

```python
BASE = Path().resolve().parent 
fund_master = pd.read_csv(BASE / "data" / "raw" / "01_fund_master.csv")
nav_history = pd.read_csv(BASE / "data" / "raw" / "02_nav_history.csv")
master_codes = set(fund_master["amfi_code"].astype(str))
nav_codes    = set(nav_history["amfi_code"].astype(str))

missing_in_nav    = master_codes - nav_codes
missing_in_master = nav_codes - master_codes

print("========== DATA QUALITY REPORT — DAY 1 ==========")
print(f"Total funds in fund_master      : {len(master_codes)}")
print(f"Unique funds in nav_history     : {len(nav_codes)}")
print(f"Codes in master BUT NOT in nav  : {missing_in_nav  or 'None — all matched ✓'}")
print(f"Codes in nav BUT NOT in master  : {missing_in_master or 'None ✓'}")
print(f"nav_history total rows          : {len(nav_history)}")
print(f"nav_history date range          : {nav_history['date'].min()} to {nav_history['date'].max()}")
print(f"Null values in nav_history      :\n{nav_history.isnull().sum()}")

report = f"""DATA QUALITY REPORT — DAY 1
============================
Total funds in fund_master      : {len(master_codes)}
Unique funds in nav_history     : {len(nav_codes)}
Codes in master BUT NOT in nav  : {missing_in_nav or 'None — all matched'}
Codes in nav BUT NOT in master  : {missing_in_master or 'None'}
nav_history total rows          : {len(nav_history)}
nav_history date range          : {nav_history['date'].min()} to {nav_history['date'].max()}
Null values in nav_history:
{nav_history.isnull().sum().to_string()}
"""
Path(BASE / "reports").mkdir(parents=True, exist_ok=True)

with open(BASE / "reports" / "data_quality_day1.txt", "w") as f:
    f.write(report)

print(f"\n✓ Report saved to {BASE / 'reports' / 'data_quality_day1.txt'}")
```
